```python
import pandas as pd
import numpy as np
import re # Nécessaire pour gérer les formats de temps comme "1h30"

# --- Début de la fonction magique de conversion de temps ---
def convertir_temps_en_minutes(colonne_de_temps):
    """
    Cette fonction prend une colonne de texte (ex: "30 min", "1h30") 
    et la convertit en une colonne de nombres (minutes).
    """
    # 1. Mise en minuscule pour uniformiser
    temps_nettoye = colonne_de_temps.astype(str).str.lower().str.strip()

    # 2. Conversion des heures en minutes (c'est la partie la plus complexe)
    def convertir_h_en_min(temps_str):
        if pd.isna(temps_str) or temps_str == 'nan':
            return np.nan
        
        minutes = 0
        temps_str = temps_str.replace(',', '.') # Gérer les décimales
        
        # Trouver un format "Xh" (ex: 1h, 2h30min, 1.5h)
        match_h = re.search(r'(\d+[\.\,]?\d*)\s*h', temps_str)
        if match_h:
            # Si on trouve "Xh", on multiplie X par 60
            minutes += float(match_h.group(1)) * 60
            # On retire la partie heure pour ne garder que les minutes
            temps_str = re.sub(r'(\d+[\.\,]?\d*)\s*h', '', temps_str)
            
        # Trouver un format "X min" ou "X" seul (ex: 30 min, 45)
        match_min = re.search(r'(\d+[\.\,]?\d*)', temps_str)
        if match_min:
            # Si on trouve un nombre, on l'ajoute comme minute
            minutes += float(match_min.group(1))

        # Si le résultat est zéro, cela signifie que la conversion a échoué (laisser NaN)
        return minutes if minutes > 0 else np.nan

    # Appliquer la conversion à toute la colonne
    return temps_nettoye.apply(convertir_h_en_min)
# --- Fin de la fonction ---
```

```python
# --- Étape A : CONVERSION (Créer les nouvelles colonnes numériques) ---

# On applique la fonction de conversion sur les colonnes "object"
dataframe['preparation_time_minutes'] = convertir_temps_en_minutes(dataframe['preparation_time'])
dataframe['cooking_time_minutes'] = convertir_temps_en_minutes(dataframe['cooking_time'])

print("Conversion des formats de texte vers les minutes numériques effectuée.")


# --- Étape B : IMPUTATION (Remplir les valeurs manquantes NaN) ---

# 1. Trouver la médiane (le temps le plus fréquent, plus stable que la moyenne)
median_prep = dataframe['preparation_time_minutes'].median()
median_cook = dataframe['cooking_time_minutes'].median()

print(f"\n-> Médiane Préparation utilisée pour l'imputation : {median_prep:.1f} minutes")
print(f"-> Médiane Cuisson utilisée pour l'imputation : {median_cook:.1f} minutes")

# 2. Remplacer les valeurs manquantes (NaN) par cette médiane
dataframe['preparation_time_cleaned'] = dataframe['preparation_time_minutes'].fillna(median_prep)
dataframe['cooking_time_cleaned'] = dataframe['cooking_time_minutes'].fillna(median_cook)

print("Remplissage des 2 valeurs manquantes par la médiane effectué.")


# --- Étape C : RECALCUL (Calculer le temps total correct) ---

dataframe['total_time_cleaned'] = dataframe['preparation_time_cleaned'] + dataframe['cooking_time_cleaned']

print("\n--- Vérification finale des types de données ---")
# On vérifie que les nouvelles colonnes sont bien FLOAT64 (numériques) et complètes.
dataframe[['preparation_time_cleaned', 'cooking_time_cleaned', 'total_time_cleaned']].info()
```
